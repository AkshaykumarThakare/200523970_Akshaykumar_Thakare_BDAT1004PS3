**Step 4. Discover what is the mean age per occupation**

```python
# Importing necessary libraries
import pandas as pd  # Importing the pandas library and using the alias 'pd'
import seaborn as sns  # Importing the seaborn library and using the alias 'sns'
import matplotlib.pyplot as plt  # Importing the pyplot module from matplotlib and using the alias 'plt'

# Replacing 'your_data_url' with the actual URL of your CSV data
data_url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'  # Specifying the URL of the CSV data
users = pd.read_csv(data_url, sep='|', index_col='user_id')  # Reading the CSV data into a pandas DataFrame

# Calculating the mean age per occupation using groupby and reset_index to obtain a DataFrame
mean_age_per_occupation = users.groupby('occupation')['age'].mean().reset_index()

# Setting the plotting style using seaborn
sns.set(style="whitegrid")

# Creating a bar plot with specified size
plt.figure(figsize=(20, 10))

# Creating a bar plot using seaborn, specifying data, and color palette
bar_plot = sns.barplot(x='occupation', y='age', data=mean_age_per_occupation, palette='deep')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adding data labels (mean ages) on top of each bar
for index, value in enumerate(mean_age_per_occupation['age']):
    bar_plot.text(index, value + 0.1, f'{value:.2f}', ha='center', va='bottom', fontsize=18)

# Setting labels and title for the plot
bar_plot.set(xlabel='Occupation', ylabel='Mean Age', title='Mean Age per Occupation')

# Displaying the plot
plt.show()

# Printing the DataFrame showing mean age per occupation
print(mean_age_per_occupation)
```

**Step 5. Discover the Male ratio per occupation and sort it from the most to the least**

```python

# Calculate the male ratio per occupation
male_ratio_per_occupation = users.groupby('occupation')['gender'].apply(lambda x: (x == 'M').mean()).reset_index()
male_ratio_per_occupation = male_ratio_per_occupation.rename(columns={'gender': 'male_ratio'})

# Sort by male ratio in descending order
male_ratio_per_occupation = male_ratio_per_occupation.sort_values(by='male_ratio', ascending=False)

# Display the result
print(male_ratio_per_occupation)
```

**Step 6. For each occupation, calculate the minimum and maximum ages**

```python
users.groupby('occupation').age.agg(['min', 'max'])
```

**Step 7. For each combination of occupation and sex, calculate the mean age**

```python
# Grouping the 'users' DataFrame by 'occupation' and 'gender' columns, then calculating the mean age for each group
users.groupby(['occupation', 'gender']).age.mean()
```

**Step 8. For each occupation present the percentage of women and men**

```python
gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'}) # Grouping the 'users' DataFrame by 'occupation' and 'gender', then aggregating the count of each gender within each occupation
occup_count = users.groupby(['occupation']).count() # Grouping users by occupation
occup_gender = gender_ocup.div(occup_count, level = "occupation") # Counting
occup_gender.loc[:, 'gender'] # Selecting Gender
```

**Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend. Don't forget to label your axes!**

```python
import pandas as pd
import matplotlib.pyplot as plt

marriage_divorce_data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

# Creatinf a line plot for marriages and divorces per capita
plt.figure(figsize=(12, 8))

# Ploting marriages per capita
plt.plot(marriage_divorce_data['Year'], marriage_divorce_data['Marriages'], label='Marriages per Capita', color='blue', marker='o')

# Ploting divorces per capita
plt.plot(marriage_divorce_data['Year'], marriage_divorce_data['Divorces'], label='Divorces per Capita', color='red', marker='o')

# Setting title and labels
plt.title('Number of Marriages and Divorces Per Capita in the U.S. (1867-2014)')
plt.xlabel('Year')
plt.ylabel('Per Capita Rate')

# Displaying legends
plt.legend()

# Adding grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Displaying the plot
plt.show()
```

# Question 8:

**Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort the actors by their kill count and label each bar with the corresponding actor's name.**

```python
hollywood_actor_kills = pd.read_csv('actor_kill_counts.csv')
print(hollywood_actor_kills.columns)
```

```python
import pandas as pd

hollywood_actor_kills = pd.read_csv('actor_kill_counts.csv')
# Sortting the DataFrame by kill count in descending order
sorted_actors_data = hollywood_actor_kills.sort_values(by='Count', ascending=True)

# Creating a horizontal bar chart for the deadliest actors
plt.figure(figsize=(10, 8))

# Plotting the horizontal bars
bars = plt.barh(sorted_actors_data['Actor'], sorted_actors_data['Count'], color='red', edgecolor='black')

# Labeling each bar with the corresponding actor's name
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f"{bar.get_width():,.0f}",
             va='center', ha='left', fontsize=10, color='black')

# Setting title and labels
plt.title('Deadliest Actors in Hollywood')
plt.xlabel('Kill Count')
plt.ylabel('Actor')

# Adding grid for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Displaying the plot
plt.show()
```

# Question 9:

**Create a pie chart showing the fraction of all Roman Emperors that were assassinated. Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.**

```python
roman_emperors = pd.read_csv('roman-emperor-reigns.csv')
print(roman_emperors.columns)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

roman_emperors = pd.read_csv('roman-emperor-reigns.csv')
# Checking if 'Cause_of_Death' contains 'assassinated' and creating a new column 'Assassinated'
roman_emperors['Assassinated'] = roman_emperors['Cause_of_Death'].str.contains('assassinated', case=False)

# Calculating the fraction of assassinated emperors
assassinated_fraction = roman_emperors['Assassinated'].mean()

# Creating a pie chart
plt.figure(figsize=(8, 8))

# Defining colors
colors = ['lightcoral', 'lightblue']

# Plotting the pie chart with labels, percentage breakdown, and other properties
plt.pie([assassinated_fraction, 1 - assassinated_fraction], labels=['Assassinated', 'Not Assassinated'],
        autopct='%1.1f%%', colors=colors, startangle=90, wedgeprops=dict(width=0.4), textprops={'fontsize': 12})

# Setting the title
plt.title('Fraction of Roman Emperors that were Assassinated', fontsize=14)

# Ensuring equal aspect ratio for a circular pie chart
plt.axis('equal')

# Displaying the plot
plt.show()
```
